```python
foo = 10
print(foo)
```

```python
foo += 10  # value is now 20 (10 + 10)
print(foo)
```

```python
foo -= 5   # value is now 15 (20 - 5)
print(foo)
```

```python
foo *= 16  # value is now 240 (15 * 16)
print(foo)
```

```python
foo //= 5  # value is now 48 (240 // 5)
print(foo)
```

```python
foo /= 4   # value is now 144.0 (48 / 4) 7
print(foo)
```

```python
foo **= 2  # value is now 144.0 (12.0 ** 2)
print(foo)
```

```python
foo %= 51  # value is now 42.0 (144.0 % 51)
print(foo)
```
